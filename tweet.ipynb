```python
import csv
from sklearn.feature_extraction.text import CountVectorizer
import numpy as np
from sklearn.feature_extraction.text import TfidfTransformer
from sklearn.pipeline import Pipeline
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import SGDClassifier
from sklearn.model_selection import GridSearchCV
```

Read the train.csv to get the traning data. After observing the training data. I decided to only extract id, text and lebal. Because other fields seems like all same and don't influence much. So I use text and label as training data. 
https://scikit-learn.org/stable/tutorial/text_analytics/working_with_text_data.html

https://scikit-learn.org/stable/modules/feature_extraction.html

```python
with open('train.csv',newline='') as csvfile:
    spamreader = csv.reader(csvfile)
    i=0
    #dic={}
    xTr = []
    yTr = []
    for row in spamreader:
        #only take id,text and label
        if i !=0:
            #dic.update({row[0]:[row[1],row[-1]]})
            xTr.append(row[1])
            yTr.append(int(row[-1]))
        i += 1
```

```python
with open('test.csv',newline='') as csvfile:
    spamreader = csv.reader(csvfile)
    i=0
    #dic={}
    xTe = []
    for row in spamreader:
        #only take id,text and label
        if i !=0:
            #dic.update({row[0]:[row[1],row[-1]]})
            xTe.append(row[1])
        i += 1
```

```python
#1-gram
# vectorizer = CountVectorizer()
# X = vectorizer.fit_transform(xTr)
#2-grams
bigram_vectorizer = CountVectorizer(ngram_range=(1,2),token_pattern=r'\b\w+\b',min_df=1)
X = bigram_vectorizer.fit_transform(xTr)
```

```python
tf_transformer = TfidfTransformer(use_idf=False).fit(X)
X_tf = tf_transformer.transform(X)
X_tf.shape
```

```python

clf = MultinomialNB().fit(X_tf, yTr)
```

```python
#NB
# text_clf = Pipeline([('vect', CountVectorizer()),('tfidf', TfidfTransformer()),('clf', MultinomialNB())])
text_clf = Pipeline([('vect', CountVectorizer()),('tfidf', TfidfTransformer()),('clf', SGDClassifier(loss='hinge', penalty='l2',
alpha=1e-3, random_state=42,n_iter=5))])
```

```python
#grid search
parameters = {'vect_ngram_range':[(1,1),(1,2)],
               'tfidf_use_idf': (True,False),
              'clf_alpha': (1e-2,1e-3)}
gs_clf = GridSearchCV(text_clf,parameters,cv=5,iid=False,n_jobs=-1)
n = int(0.8*len(yTr))
print(n)
gs_clf = gs_clf.fit(xTr[:n], yTr[:n])
```

```python
text_clf.fit(xTr,yTr)
```

```python
predicted = text_clf.predict(xTe)
print(predicted)
```

```python
ID = np.arange(0,len(predicted)+1)
data = zip(ID,predicted)
with open('result_svm_bigram.csv','w',newline='') as csvfile:
    spamwriter = csv.writer(csvfile)
    spamwriter.writerow(["ID","Label"])
    for d in data:
        spamwriter.writerow(d)
```
